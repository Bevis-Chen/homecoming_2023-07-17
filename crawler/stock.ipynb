```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url= "https://tw.stock.yahoo.com/"
resp= requests.get(url)
resp
```

```python
soup= BeautifulSoup(resp.text, "lxml")
soup
```

```python
main= soup.find(class_="P(0) M(0)").find_all("li")
main
```

```python
main[0].find(class_= "Fx(n) Fw(b) Fz(16px)").text.strip()
```

```python
main[0].find(class_= "Fw(600) Fz(16px) D(f) Ai(c) C($c-trend-down)").text.strip()
```
